```python
data = pd.read_csv("battles.csv")
```

```python
location = data["location"].value_counts().to_dict()
battles_locations = location.keys()
battles_locations_counts = location.values()
```

```python
battles_locations_counts
```

```python
len(battles_locations)
```

```python
explode = [0.1 if location == "Riverrun" or location == "Storm's End" or location == "Golden Tooth" else 0 for location in battles_locations]
```

```python
plt.title("Game Of Thrones Battels Locations", pad=10, fontsize=100, color="Goldenrod",loc="left",fontfamily='fantasy')
plt.rcParams.update({"figure.figsize":(40,40),"font.size":25})
shades = np.linspace(0, 1, 27)
colors = [(0, 0, shade) for shade in shades]
wedges, labels, autopcts = plt.pie(battles_locations_counts,autopct="%1.1f%%",startangle=0,explode=explode,labels=battles_locations,shadow=True,colors=colors,textprops={'color': 'white'})
for label in labels:
    label.set_color('black')
plt.legend(loc='best', fontsize=0.5)
plt.show()
```

```python
attacker = data["attacker_1"].value_counts().to_dict()
attacker_name = attacker.keys()
attacker_counts = attacker.values()
```

```python
explode1 = [0.02 if location == "Lannister" or location == "Baratheon" else 0 for location in attacker_name]
```

```python
plt.title("Game Of Thrones Attackers", pad=10, fontsize=100, color="Goldenrod",loc="left",fontfamily='fantasy')
plt.rcParams.update({"figure.figsize":(40,40),"font.size":25})
shades = np.linspace(0, 1, 27)
colors = [(shade, 0, 0) for shade in shades]
wedges, labels, autopcts = plt.pie(attacker_counts,autopct="%1.1f%%",startangle=0,explode=explode1,labels=attacker_name,shadow=True,colors=colors,textprops={'color': 'white'})
for label in labels:
    label.set_color('black')
plt.legend(loc='best', fontsize=0.5)
plt.show()
```
